### Import libraries

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

from collections import Counter

sns.set(style="whitegrid")
sns.set_palette("pastel")
```

### Read json

Read the generated .json file during the crawling process of the URL's:
- url_crawling\url_crawling\spiders\products\products.json

```python
JSON_PATH = r"url_crawling\url_crawling\spiders\products\products_file.json"

# Read JSON and filter out URLs ending with "/"
df = pd.read_json(JSON_PATH)
df = df[~df['url'].str.endswith('/')].reset_index(drop=True)
```

### Explore the extracted json

Create two separate dataframes: one for failed URLs and another for extracted products.

```python
df_errors = df[df['error'].notna()].drop(columns='product_name').reset_index(drop=True)
df_products = df[df['error'].isna()].drop(columns='error').reset_index(drop=True)
```

Let's start with something generic by checking the number of errors and the number of extract products

```python
print(f"We have {df_errors.shape[0]} failed URLs and {df_products.shape[0]} extracted products out of a total of {df.shape[0]} URLs.")
```

Let's have a look at the causes of the failed URL's

```python
# Update elements containing 'DNS lookup failed: no results for hostname lookup'
df_errors['error'] = df_errors['error'].apply(lambda error: 'DNS lookup failed: no results for hostname lookup' if 'DNS lookup failed: no results for hostname lookup' in error else error)

# Define a dictionary mapping full error messages to categories
error_categorization = {
    'No product name found': 'Extraction Error',
    'Ignoring non-200 response': 'HTTP Error',
    'DNS lookup failed: no results for hostname lookup': 'DNS Error'
}

# Apply the categorization to your updated list of errors
errors_list = df_errors['error'].map(error_categorization).fillna('Other Error').values.tolist()

# Count each categorized error
errors_counts = Counter(errors_list).most_common()
top_5_errors_counts = errors_counts[:4]
```

Plot the top 3 errors we encountered

```python
# Extract the error categories and counts for plotting
error_categories = [error for error, count in top_5_errors_counts]
error_counts = [count for error, count in top_5_errors_counts]

# Plot the bar chart
plt.figure(figsize=(8, 4))
ax = sns.barplot(x=error_categories, y=error_counts)
ax.set_title('HTTP Status Codes for Failed URLs')
plt.xticks(rotation=45)
plt.show()
```
